Triky

```python
#Triky o juego del gato

def imprimir_tablero(tablero):
    for fila in tablero:
        print(" | ".join(fila))
        print("-" * 9)

def verificar_ganador(tablero, jugador):
    # Verifica filas, columnas y diagonales
    for i in range(3):
        if all([celda == jugador for celda in tablero[i]]):  # filas
            return True
        if all([tablero[j][i] == jugador for j in range(3)]):  # columnas
            return True
    if all([tablero[i][i] == jugador for i in range(3)]) or \
       all([tablero[i][2 - i] == jugador for i in range(3)]):  # diagonales
        return True
    return False

def tablero_lleno(tablero):
    return all(cell != " " for fila in tablero for cell in fila)

def jugar_triky():
    tablero = [[" " for _ in range(3)] for _ in range(3)]
    jugador_actual = "X"

    while True:
        print(f"\nTurno del jugador {jugador_actual}")
        imprimir_tablero(tablero)

        try:
            fila = int(input("Ingresa la fila (0-2): "))
            columna = int(input("Ingresa la columna (0-2): "))
        except ValueError:
            print("Entrada inválida. Usa números del 0 al 2.")
            continue

        if 0 <= fila <= 2 and 0 <= columna <= 2:
            if tablero[fila][columna] == " ":
                tablero[fila][columna] = jugador_actual
                if verificar_ganador(tablero, jugador_actual):
                    imprimir_tablero(tablero)
                    print(f"¡El jugador {jugador_actual} ha ganado!")
                    break
                elif tablero_lleno(tablero):
                    imprimir_tablero(tablero)
                    print("¡Empate!")
                    break
                jugador_actual = "O" if jugador_actual == "X" else "X"
            else:
                print("Esa celda ya está ocupada.")
        else:
            print("Coordenadas fuera de rango. Usa números del 0 al 2.")

# Ejecutar el juego
jugar_triky()
```

```python
#Gráfico 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Cargar datos
url = "https://raw.githubusercontent.com/JuanTovar12/Seminario/refs/heads/main/salida_mensual_pasajeros_aeropuerto_destino_internacional.csv"
df = pd.read_csv(url)


# Convertir a fecha
df["fecha"] = pd.to_datetime(df["sal_periodo"].astype(str), format="%Y%m")


# Agrupar por fecha
viajes_mensuales = df.groupby("fecha")["sal_valor"].sum().reset_index()


# Datos base
x = viajes_mensuales["fecha"]
y = viajes_mensuales["sal_valor"]


# Crear gráfico
plt.figure(figsize=(14, 6))
plt.plot(x, y, label="Pasajeros", color="blue", marker='o')


# Pintar áreas debajo: verde si sube, rojo si baja
for i in range(1, len(y)):
    if y[i] >= y[i-1]:
        plt.fill_between(x[i-1:i+1], y[i-1:i+1], color="green", alpha=0.3)
    else:
        plt.fill_between(x[i-1:i+1], y[i-1:i+1], color="red", alpha=0.3)


# Estética
plt.title("Tendencia mensual de pasajeros internacionales")
plt.xlabel("Fecha")
plt.ylabel("Número de pasajeros")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

#Gráfico 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Cargar datos
url = "https://raw.githubusercontent.com/JuanTovar12/Seminario/refs/heads/main/salida_mensual_pasajeros_aeropuerto_destino_internacional.csv"
df = pd.read_csv(url)


# Convertir a fecha
df["fecha"] = pd.to_datetime(df["sal_periodo"].astype(str), format="%Y%m")


# Agrupar por fecha
viajes_mensuales = df.groupby("fecha")["sal_valor"].sum().reset_index()


# Datos base
x = viajes_mensuales["fecha"]
y = viajes_mensuales["sal_valor"]


# Crear gráfico
plt.figure(figsize=(14, 6))
plt.plot(x, y, label="Pasajeros", color="blue", marker='o')


# Pintar áreas debajo: verde si sube, rojo si baja
for i in range(1, len(y)):
    if y[i] >= y[i-1]:
        plt.fill_between(x[i-1:i+1], y[i-1:i+1], color="green", alpha=0.3)
    else:
        plt.fill_between(x[i-1:i+1], y[i-1:i+1], color="red", alpha=0.3)


# Estética
plt.title("Tendencia mensual de pasajeros internacionales")
plt.xlabel("Fecha")
plt.ylabel("Número de pasajeros")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()






#Gráfico 2
# Obtener los 6 países más visitados
top_paises = df.groupby("sal_destinoint")["sal_valor"].sum().nlargest(6).index.tolist()
df_top6 = df[df["sal_destinoint"].isin(top_paises)]


# Graficar comportamiento de cada país individualmente
plt.figure(figsize=(18, 12))
for i, pais in enumerate(top_paises):
    plt.subplot(3, 2, i + 1)
    datos_pais = df_top6[df_top6["sal_destinoint"] == pais]
    datos_pais = datos_pais.groupby("fecha")["sal_valor"].sum().reset_index()
    sns.lineplot(data=datos_pais, x="fecha", y="sal_valor", marker='o', color='orange')
    plt.title(f"Pasajeros hacia {pais}")
    plt.xlabel("Fecha")
    plt.ylabel("Número de pasajeros")
    plt.xticks(rotation=45)
    plt.grid(True)


plt.tight_layout()
plt.show()




#Gráfico 3
# Agrupar por fecha y país
df_top6_grouped = df_top6.groupby(["fecha", "sal_destinoint"])["sal_valor"].sum().reset_index()


# Gráfico conjunto
plt.figure(figsize=(14, 7))
sns.lineplot(data=df_top6_grouped, x="fecha", y="sal_valor", hue="sal_destinoint", marker="o")
plt.title("Comparativo de pasajeros a los 6 países más visitados")
plt.xlabel("Fecha")
plt.ylabel("Número de pasajeros")
plt.legend(title="País de destino")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


#Gráfico 4
# Filtrar el país más visitado (por ejemplo Estados Unidos)
pais_mas_visitado = top_paises[0]
df_pais = df[df["sal_destinoint"] == pais_mas_visitado]
df_pais = df_pais.groupby("fecha")["sal_valor"].sum().reset_index()


# Gráfico
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_pais, x="fecha", y="sal_valor", marker='o', color='green')
plt.title(f"Comportamiento de pasajeros hacia {pais_mas_visitado}")
plt.xlabel("Fecha")
plt.ylabel("Número de pasajeros")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Estadísticas generales
print("Análisis del país más visitado:", pais_mas_visitado)
print(df_pais.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

aeropuesto_df = pd.read_csv("https://raw.githubusercontent.com/Samael-hc/Seminario_prueba/refs/heads/main/salida_mensual_pasajeros_aeropuerto_destino_internacional.csv")

column_rename_map = {
    "sal_codigo": "codio.head()",
    "sal_destinoint": "pais_destino",
    "sal_indicador": "indicador_salida",
    "sal_periodo": "periodo",
    "sal_valor": "valor_salida"
}

aeropuerto_df_renombrado = aeropuesto_df.rename(columns=column_rename_map)

aeropuerto_df_renombrado.head()
```

```python
us_data = aeropuerto_df_renombrado[aeropuerto_df_renombrado['codigo_pais'] == 'US'].copy()
us_data.loc[:, 'periodo'] = pd.to_datetime(us_data['periodo'], format='%Y%m')
us_data['año'] = us_data['periodo'].dt.year
us_data['mes'] = us_data['periodo'].dt.month
mejor_mes_por_año = us_data.groupby(['año', 'mes'])['valor_salida'].sum().reset_index()
mejor_mes_por_año = mejor_mes_por_año.loc[mejor_mes_por_año.groupby('año')['valor_salida'].idxmax()]
print(mejor_mes_por_año)

plt.figure(figsize=(14, 8))
plt.plot(mejor_mes_por_año['año'].astype(str) + '-' + mejor_mes_por_año['mes'].astype(str),
         mejor_mes_por_año['valor_salida'], marker='o', linestyle='-', color='b')
plt.title('Mejor mes de cada año para US en términos de valor_salida')
plt.xlabel('Año-Mes')
plt.ylabel('Valor de Salida')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

moda_mes = mejor_mes_por_año['mes'].mode()[0]
print(f"La moda del mes es: {moda_mes}")
frecuencia_meses = mejor_mes_por_año['mes'].value_counts()

plt.figure(figsize=(10, 6))
frecuencia_meses.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Frecuencia de los mejores meses en términos de valor_salida')
plt.xlabel('Mes')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
